```python
movie= [
    "The Shawshank Redemption",
    "The Dark Knight",
    "Inception",
    "Pulp Fiction",
    "The Matrix",
    "Interstellar",
    "Parasite",
    "Gladiator",
    "Titanic",
    "The Lord of the Rings: The Fellowship of the Ring"
]
year= [
    1994,
    2008,
    2010,
    1994,
    1999,
    2014,
    2019,
    2000,
    1997,
    2001
]
is_favorite= [
    True,
    False,
    True,
    False,
    True,
    False,
    True,
    False,
    True,
    False
]

def print_favorite_movies(movies, years, favorites):
    for movie, year, favorite in zip(movies, years, favorites):
        if favorite:
            print(f"{movie} ({year}) is one of my favorite movies.")
        else:
            print(f"{movie} ({year}) is not one of my favorite movies.")

print_favorite_movies(movie, year, is_favorite)
```
