Calculadora sencilla

```python
num1 = float(input("Ingresa el primer número: "))
num2 = float(input("Ingresa el segundo número: "))

def suma(num1, num2):
    resultado = num1 + num2
    return f"El resultado es: {resultado}"

def res(num1, num2):
    resultado = num1 - num2
    return f"El resultado es: {resultado}"

def mult(num1, num2):
    resultado = num1 * num2
    return f"El resultado es: {resultado}"

def div(num1, num2):
    if num2 != 0:
        resultado = num1 / num2
        return f"El resultado es: {resultado}"
    else:
        return "No es posible dividir por 0"

operation = input("""
Elija la operación:

Suma: 1
Resta: 2
Multiplicación: 3
División: 4

Opción:
""")

match operation:
    case "1":
        print(suma(num1, num2))
    case "2":
        print(res(num1, num2))
    case "3":
        print(mult(num1, num2))
    case "4":
        print(div(num1, num2))
    case _:
        print("Opción inválida")
```
